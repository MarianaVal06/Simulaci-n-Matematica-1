Existe una función que recibe un número entero y te dice si es un número primo.

```python
import numpy as np
def primo(x):
    c=np.floor(np.sqrt(x))
    i=2
    while (i<=c) :
        if x%i==0:
            return("No es primo")
        i += 1
    return("Es primo")    
```

Existe una función que recibe un número y te regresa el factorial de ese número.

```python
def factorial(n):
    factorial = 1
    for i in range(1,n + 1):
        factorial *= i
    return factorial
```
